# Import Library & Load Dataset

```python
import pandas as pd

# URL of the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Read and save CSV
df = pd.read_csv(url)
df.to_csv("titanic.csv", index=False)  # Save locally

print("Download completed & Titanic dataset saved as 'titanic.csv' ")
```

# 7️⃣ Find the average age of passengers in each class (Pclass).

```python
average_age_per_class = df.groupby("Pclass")["Age"].mean()
print(average_age_per_class)

# => groupby("Pclass") → Pclass (1st, 2nd, 3rd class) bases.
# => ["Age"].mean() → each class's Age value calculate will be  (average) .
```

# 8️⃣ Count how many passengers survived (Survived == 1) in each class.

```python
survivors_per_class = df.groupby("Pclass")["Survived"].sum()
print(survivors_per_class)
```

# 9️⃣ Find out the youngest and oldest passenger on the Titanic.

```python
# for find youngest passenger
youngest_age = df["Age"].min()

# for Find the oldest passenger
oldest_age = df["Age"].max()

# Show results
print(f"Youngest Passenger Age: {youngest_age}")
print(f"Oldest Passenger Age: {oldest_age}")
```

# 🔟 Create a new column "FamilySize" = SibSp + Parch + 1, then analyze its impact on survival.

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1  # +1 includes the passenger themselves

# Analyze impact on survival
survival_by_family_size = df.groupby("FamilySize")["Survived"].mean()

# Show result
print(survival_by_family_size)
```
